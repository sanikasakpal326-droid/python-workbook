# Q.1) Write a program to find the maximum and minimum element in a tuple without using max() and min() functions. 

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

max_value = t[0]
min_value = t[0]

for element in t:
    if element > max_value:
        max_value = element
    if element < min_value:
        min_value = element

print("Maximum element:", max_value)
print("Minimum element:", min_value)
```

# Q.2) Write a program to count the occurrence of each element in a tuple. 
#Example: For tuple (1, 2, 3, 2, 1, 3, 2), output should show 1:2, 2:3, 3:2 

```python
t = (4, 6, 2, 8, 1)
i = 0
max_value = t[0]
min_value = t[0]

while i < len(t):
    if t[i] > max_value:
        max_value = t[i]
    if t[i] < min_value:
        min_value = t[i]
    i += 1

print("Maximum:", max_value)
print("Minimum:", min_value)
```

# Q.3) Write a program to convert a tuple of strings into a single string. 
#Example: ("Hello", "World", "Python") should become "HelloWorldPython"

```python
t = ("Hello", "World", "Python")

result = ""

for word in t:
    result = result + word

print("Single string:", result)
```

# Q.4) Write a program to find the sum of all even numbers in a tuple.

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

sum_even = 0

for num in t:
    if num % 2 == 0:
        sum_even = sum_even + num

print("Sum of even numbers:", sum_even)
```

# Q.5) Write a program to remove duplicates from a tuple (return a new tuple with unique elements). 

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

unique = []

for element in t:
    if element not in unique:
        unique.append(element)

result = tuple(unique)

print("Tuple without duplicates:", result)
```

# Q.1) Write a program to find the second largest element in a tuple. 

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

max_val = t[0]
second_max = -10**9   

for num in t:
    if num > max_val:
        second_max = max_val
        max_val = num
    elif num != max_val and num > second_max:
        second_max = num

print("Second largest element:", second_max)
```

# Q.2) Write a program to sort a tuple of tuples based on the second element. 
Example: ((1, 5), (3, 2), (2, 8)) should become ((3, 2), (1, 5), (2, 8)) 

```python
t = ((1, 5), (3, 2), (2, 8))

sorted_tuple = tuple(sorted(t, key=lambda x: x[1]))

print("Sorted tuple:", sorted_tuple)
```

# Q.3) Write a program to check if all elements in a tuple are unique. 

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

is_unique = True

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] == t[j]:
            is_unique = False
            break

if is_unique:
    print("All elements in the tuple are unique.")
else:
    print("The tuple contains duplicate elements.")
```

# Q.4) Write a program to swap the first and last element of a tuple.

```python
t = tuple(map(int, input("Enter elements of the tuple: ").split()))

if len(t) < 2:
    print("Tuple must have at least two elements")
else:
    swapped = (t[-1],) + t[1:-1] + (t[0],)
    print("Tuple after swapping:", swapped)
```

# Q.1) Write a program to find common elements between two tuples.

```python
t1 = tuple(map(int, input("Enter elements of first tuple: ").split()))
t2 = tuple(map(int, input("Enter elements of second tuple: ").split()))

common = []

for element in t1:
    if element in t2 and element not in common:
        common.append(element)

result = tuple(common)

print("Common elements:", result)
```

# Q.2) Write a program to create a tuple of squares of numbers from 1 to n. 
Example: If n=5, output should be (1, 4, 9, 16, 25) 

```python
n = int(input("Enter the value of n: "))

squares = ()

for i in range(1, n + 1):
    squares = squares + (i * i,)

print("Tuple of squares:", squares)
```
