# Question Set A

# Q.1) Write a program to find the union, intersection, and difference of two sets entered by the user. 

```python
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

union_set = set1 | set2
intersection_set = set1 & set2
difference_set = set1 - set2

print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (Set1 - Set2):", difference_set)
```

# Q.2) Write a program to check if a set is a subset of another set. 

```python
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

if set1.issubset(set2):
    print("First set is a subset of the second set.")
else:
    print("First set is not a subset of the second set.")
```

# Q.3) Write a program to remove duplicates from a list using sets.

```python
lst = [1, 2, 3, 2, 4, 1]
unique = []
seen = set()

for item in lst:
    if item not in seen:
        unique.append(item)
        seen.add(item)

print(unique)
```

```python
lst = list(map(int, input("Enter elements of the list: ").split()))

unique_list = list(set(lst))

print("List after removing duplicates:", unique_list)
```

# Q.4) Write a program to find elements that are present in the first set but not in the second set.

```python
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

result = set1 - set2

print("Elements present in first set but not in second set:", result)
```

# Q.5) Write a program to check if two sets are disjoint (have no common elements).

```python
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

if set1.isdisjoint(set2):
    print("The two sets are disjoint.")
else:
    print("The two sets are not disjoint.")
```

# Question Set B

# Q.1) Write a program to find symmetric difference between two sets. 

```python
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

result = set1 ^ set2

print("Symmetric Difference:", result)
```

# Q.2) Write a program to add multiple elements to a set at once from user input. 

```python
s = set()

elements = input("Enter elements to add to the set: ").split()

s.update(elements)

print("Updated set:", s)
```

# Q.3) Write a program to find the length of a set and check if a specific element exists in it. 

```python
s = set(map(int, input("Enter elements of the set: ").split()))

print("Length of the set:", len(s))

element = int(input("Enter element to check: "))

if element in s:
    print("Element exists in the set.")
else:
    print("Element does not exist in the set.")
```

# Q.1) Write a program that takes two sentences as input and finds: 
● Common words in both sentences 
● Words unique to the first sentence 
● Words unique to the second sentence

```python
sentence1 = input("Enter first sentence: ")
sentence2 = input("Enter second sentence: ")

set1 = set(sentence1.split())
set2 = set(sentence2.split())

common_words = set1 & set2
unique_first = set1 - set2
unique_second = set2 - set1

print("Common words:", common_words)
print("Words unique to first sentence:", unique_first)
print("Words unique to second sentence:", unique_second)
```

# Q.2) Write a program to perform all set operations (union, intersection, difference, symmetric difference) on three sets.


```python
# Input three sets
set1 = set(input("Enter elements of Set 1 (space-separated): ").split())
set2 = set(input("Enter elements of Set 2 (space-separated): ").split())
set3 = set(input("Enter elements of Set 3 (space-separated): ").split())

# Union of three sets
union_set = set1 | set2 | set3
print("\nUnion of sets:", union_set)

# Intersection of three sets
intersection_set = set1 & set2 & set3
print("Intersection of sets:", intersection_set)

# Difference operations
print("Set1 - Set2 - Set3:", set1 - set2 - set3)
print("Set2 - Set1 - Set3:", set2 - set1 - set3)
print("Set3 - Set1 - Set2:", set3 - set1 - set2)

# Symmetric Difference of three sets
symmetric_difference = set1 ^ set2 ^ set3
print("Symmetric Difference of sets:", symmetric_difference)
```
