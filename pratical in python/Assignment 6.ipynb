# Q.1) Write a function to find the maximum of three numbers.


```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

maximum = find_max(num1, num2, num3)

print("The maximum number is:", maximum)
```

# Q.2) Function to Check Whether a Number is Prime.

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print("The number is Prime")
else:
    print("The number is Not Prime")
```

# Q.3) Write a function to calculate the factorial of a number (both iterative and recursive). 

```python
def factorial_iterative(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact

num = int(input("Enter a number: "))
print("Factorial (Iterative):", factorial_iterative(num))
```

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

num = int(input("Enter a number: "))
print("Factorial (Recursive):", factorial_recursive(num))
```

# Q.4) Write a function that accepts a list and returns a new list with unique elements. 

```python
def unique_elements(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

numbers = [1, 2, 2, 3, 4, 4, 5]

print("Unique elements:", unique_elements(numbers))
```

# Q.5) Write a function to calculate the sum of all even numbers in a list. 


```python
def sum_of_even(lst):
    total = 0
    for num in lst:
        if num % 2 == 0:
            total += num
    return total

numbers = [1, 2, 3, 4, 5, 6]

print("Sum of even numbers:", sum_of_even(numbers))
```

# 1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa. 

```python
def temperature_convert(value, scale):
    if scale == 'C':
        return (value * 9/5) + 32
    elif scale == 'F':
        return (value - 32) * 5/9
    else:
        return "Invalid scale"

temp = float(input("Enter temperature value: "))
choice = input("Enter C for Celsius or F for Fahrenheit: ")

result = temperature_convert(temp, choice)
print("Converted temperature:", result)
```

# Q.2) Write a function that accepts a string and returns the count of vowels and consonants. 


```python
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0

    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

text = input("Enter a string: ")

v, c = count_vowels_consonants(text)
print("Vowels:", v)
print("Consonants:", c)
```

# Q.3) Write a function that accepts variable number of arguments and returns their average. 

```python
def average(*args):
    total = 0
    for num in args:
        total += num
    return total / len(args)

print("Average:", average(10, 20, 30, 40))
```

# Q.4) Write a lambda function to find the maximum of three numbers. 

```python
maximum = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)

print("Maximum number:", maximum(10, 25, 15))
```

# Q.1)Write a recursive function to calculate the sum of digits of a number

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return (n % 10) + sum_of_digits(n // 10)

num = int(input("Enter a number: "))

print("Sum of digits:", sum_of_digits(num))
```

# Q.2) Write a function that accepts a list of numbers and returns a dictionary with count of even and odd numbers. 
#Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3}

```python
def count_even_odd(numbers):
    result = {'even': 0, 'odd': 0}

    for num in numbers:
        if num % 2 == 0:
            result['even'] += 1
        else:
            result['odd'] += 1

    return result

nums = [1, 2, 3, 4, 5]

print(count_even_odd(nums))
```
