# Q.1) Create a class called Rectangle with attributes length and width. Add methods to calculate area and perimeter.


```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Example
r = Rectangle(10, 5)
print("Area:", r.area())
print("Perimeter:", r.perimeter())
```

# Q.2) Create a class Employee with class method to keep count of total employees. 

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def display_balance(self):
        print("Current Balance:", self.balance)
        
# example
acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
acc.display_balance()
```

# Q.5) Create a class Circle with method to calculate area. Overload the + operator to add two circles (add their radii)

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def __add__(self, other):
        return Circle(self.radius + other.radius)

c1 = Circle(3)
c2 = Circle(4)
c3 = c1 + c2

print("New Radius:", c3.radius)
```

# Q.1) Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance. 

```python
# taking every class
class Animal:
    def speak(self):
        print("Animal makes sound")

class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

class Dog(Mammal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()
d.walk()
d.bark()
```

# Q.2) Create a class Employee with class method to keep count of total employees. 


```python
class Employee:
    count = 0

    def __init__(self, name):
        self.name = name
        Employee.count += 1

    @classmethod
    def total_employees(cls):
        return cls.count

# input
e1 = Employee("Amit")
e2 = Employee("Neha")

print("Total Employees:", Employee.total_employees())
```

# Q.3) Create a class Book with special methods __str__, __len__, and __eq__. 


```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return self.title

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.pages == other.pages

b1 = Book("Python", 300)
b2 = Book("Java", 300)
# example
print(b1)
print(len(b1))
print(b1 == b2)
```

# Q.4) Create a class Temperature with property decorators for Celsius and Fahrenheit conversion. 

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

t = Temperature(25)
print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)
```

# Q.1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle, Circle, and Triangle that implement the area method.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Triangle(Shape):
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * self.b * self.h

shapes = [Rectangle(4,5), Circle(3), Triangle(4,6)]
for s in shapes:
    print("Area:", s.area())
```

# Q.2) Create a class Library that manages a collection of books. Implement methods to add books, remove books, search books, and display all books.

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)

    def search_book(self, book):
        return book in self.books

    def display_books(self):
        print("Books in Library:")
        for book in self.books:
            print(book)

lib = Library()
lib.add_book("ice breaker")
lib.add_book("verity")
lib.display_books()
```
