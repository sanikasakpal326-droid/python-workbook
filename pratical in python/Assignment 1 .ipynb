# Q.2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the values entered are duplicates otherwise it prints "ALL
#Read 6 integer values from the user

```python
nums = []
for i in range(6):
    value = int(input("Enter an integer: "))
    nums.append(value)

if len(nums) != len(set(nums)):
    print("DUPLICATES") 
else:
    print("ALL UNIQUE")
```

# Q.4) Write a program which finds sum of digits of a number.
#Example: n=130 then output is 4 (1+3+0).


```python
n = int(input("Enter a number: "))
sum_digits = 0

while n > 0:
    digits = n % 10
    sum_digits += digits
    n = n //10

print("Sum of digits is:", sum_digits)
```

# Q.5) Write a program which prints Fibonacci series of a number. 


```python
def fibonacci(n):
    a = 0
    b = 1

    print("Fibonacci series:")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

# Input from user
num = int(input("Enter the number of terms: "))

fibonacci(num)
```

# Q.1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'. 


```python
n = int(input("Enter a number: "))

print("Prime numbers till", n, "are:")

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```

# Q.4) Write a program to reverse a given number. 


```python
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = (reverse * 10) + digit
    num = num // 10

print("Reversed number:", reverse)
```

# Q.1) Write a Sequential search function which searches an item in a sorted list. The function should return the index of element to be searched in the list.

```python
def sequential_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Example sorted list
numbers = [2, 4, 6, 8, 10, 12, 14]

# Element to search
search_item = int(input("Enter element to search: "))

result = sequential_search(numbers, search_item)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found in the list")
```
