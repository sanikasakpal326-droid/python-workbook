# Q.1)write a program to create a dictionary with roll number as key and name as value for 5 students.Display all key-value pairs.

```python
students = {}
for i in range(1, 6):
    roll_no = int(input(f"Enter roll number of student {i}: "))
    name = input(f"Enter name of student {i}: ")
    students[roll_no] = name
print("\nStudent Details:")
for roll_no, name in students.items():
    print("Roll Number:", roll_no, "Name:", name)
```

# Q.2)write a program to count the frequency of each word in a given sentence using a dictionary.
#example: "helllo world hello" should give {'hello':2,'world': 1}

```python
sentence = input("Enter a sentence: ")
words = sentence.lower().split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("Word Frequency:", word_count)
```

# Q.3)write a program to merge two dictionaries entered by the user.

```python
# Input first dictionary
dict1 = {}
n1 = int(input("Enter number of elements in first dictionary: "))

for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

# Input second dictionary
dict2 = {}
n2 = int(input("\nEnter number of elements in second dictionary: "))

for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

# Merge dictionaries
merged_dict = dict1.copy()
merged_dict.update(dict2)

# Display result
print("\nFirst Dictionary:", dict1)
print("Second Dictionary:", dict2)
print("Merged Dictionary:", merged_dict)
```

# Q.4) write a program to find the key with the maximum value in a dictionary.

```python
data = {
    'A': 45,
    'B': 78,
    'C': 32,
    'D': 90,
    'E': 67
}
max_key = None
max_value = 0

for key, value in data.items():
    if value > max_value:
        max_value = value
        max_key = key

print("Key with maximum value:", max_key)
print("Maximum value:", max_value)
```

# Q.5)write a program to check if a given key exists in a dictionary.

```python
student = {
    'roll_no': 101,
    'name': 'Rahul',
    'age': 20,
    'course': 'BSc'
}
key = input("Enter the key to check: ")

if key in student:
    print("The key exists in the dictionary.")
else:
    print("The key does not exist in the dictionary.")
```

# Q.1)write a program to remove a key from a dictionary and handle the case when key doesn't exit.

```python
student = {
    'roll_no': 101,
    'name': 'Rahul',
    'age': 20,
    'course': 'BSc'
}

key = input("Enter the key to remove: ")

if key in student:
    del student[key]
    print("Key removed successfully.")
    print("Updated dictionary:", student)
else:
    print("Key does not exist in the dictionary.")
```

# Q.2) write a program to sort a dictionary by its values in descending order.

```python
marks = {
    'Maths': 85,
    'Science': 92,
    'English': 78,
    'History': 88
}

sorted_dict = dict(sorted(marks.items(), key=lambda item: item[1], reverse=True))

print("Dictionary sorted by values (Descending Order):")
print(sorted_dict)
```

# Q.3) write a program to create a dictonary from two lists- one containing keys and another containing values

```python
keys = ['roll_no', 'name', 'age', 'course']
values = [101, 'Rahul', 20, 'BSc']

student = dict(zip(keys, values))

print("Created Dictionary:")
print(student)
```

# Q.1) write a program to find all keys with the same value in a dictionary.

```python
data = {
    'A': 10,
    'B': 20,
    'C': 10,
    'D': 30,
    'E': 20
}
same_value_keys = {}

for key, value in data.items():
    if value in same_value_keys:
        same_value_keys[value].append(key)
    else:
        same_value_keys[value] = [key]

print("Keys with the same values:")
for value, keys in same_value_keys.items():
    if len(keys) > 1:
        print(f"Value {value} -> Keys {keys}")
```

# Q.2) write a program to create a dictionary where keys are numbers from 1 to n and values are their cubes.

```python
n = int(input("Enter the value of n: "))

cube_dict = {}

for i in range(1, n + 1):
    cube_dict[i] = i ** 3

print("Dictionary with cubes:")
print(cube_dict)
```
