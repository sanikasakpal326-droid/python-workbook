# Q.1) Write a program to count the number of vowels in a given string.

```python
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("Number of vowels:", count)
```

# Q.2) Write a program to check if a string is a palindrome (reads the same forwards and backwards).

#Example: "madam" is a palindrome, "python" is not.

```python
a = input("Enter a string: ")
rev = ""
for ch in a:
    rev = ch + rev

if a == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# Q.3) Write a program to reverse a string without using built-in reverse functions. 

```python
b = input("Enter a string: ")
rev = ""

for ch in b:
    rev = ch + rev

print("Reversed string:", rev)
```

# Q.4) Write a program to count the frequency of each character in a string. 
#Example: For "hello", output should be h:1, e:1, l:2, o:1 

```python
s = input("Enter a string: ")
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for key in freq:
    print(key, ":", freq[key])
```

# Q.5) Write a program to remove all whitespace from a string. 

```python
s = input("Enter a string: ")
result = ""

for ch in s:
    if ch != " ":
        result += ch

print("String without whitespace:", result)
```

# Q.1) Write a program to find the first non-repeating character in a string.

```python
s = input("Enter a string: ")

for ch in s:
    if s.count(ch) == 1:
        print("First non-repeating character:", ch)
        break
else:
    print("No non-repeating character found")
```

# Q.2) Write a program to check if two strings are anagrams (contain same characters in different order). 
#Example: "listen" and "silent" are anagrams. 

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not Anagrams")
```

# Q.3) Write a program to convert a string to title case without using the title() method. 
#Example: "python programming" should become "Python Programming" 

```python
s = input("Enter a string: ")
words = s.split()
result = ""

for w in words:
    result += w[0].upper() + w[1:] + " "

print(result.strip())
```

# Q.4) Write a program to replace all occurrences of a substring with another substring in a given string.

```python
s = input("Enter a string: ")
old = input("Enter substring to replace: ")
new = input("Enter new substring: ")

result = s.replace(old, new)
print("Modified string:", result)
```

# Q.1) Write a program to find the longest word in a sentence. 

```python
sentence = input("Enter a sentence: ")

words = sentence.split() 
longest_word = ""

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print("Longest word is:", longest_word)
```

# Q.2) ) Write a program to check if a string contains only digits, only alphabets, or both. 

```python
string = input("Enter a string: ")

if string.isdigit():
    print("The string contains only digits.")
elif string.isalpha():
    print("The string contains only alphabets.")
elif string.isalnum():
    print("The string contains both alphabets and digits.")
else:
    print("The string contains special characters.")
```
